```python
import matplotlib.pyplot as plt
import numpy as np

# Données pour Nyström et RFF
accuracy_nystrom = [79, 79, 80, 82, 87]  # Accuracy values for Nyström
accuracy_rff = [79, 79, 79, 81, 84]      # Accuracy values for RFF

# Temps et mémoire pour Nyström
nystrom_time = [0, 0, 2, 28, 136]
nystrom_memory = [17, 9, 43, 86, 627]

# Temps et mémoire pour RFF
rff_time = [0, 0, 0, 5, 16]
rff_memory = [0, 0, 0, 71, 179]

# Ajouter les données du modèle de base (accuracy = 87%)
baseline_accuracy = 87
baseline_time = 56
baseline_memory = 161

# Création du graphe
fig, ax1 = plt.subplots(figsize=(10, 6))

# Axe gauche : Temps
ax1.set_xlabel("Accuracy (%)")
ax1.set_ylabel("Clock Time (s)", color='tab:blue')
ax1.plot(accuracy_nystrom, nystrom_time, label="Nyström (Time)", color='tab:blue', marker='o')
ax1.plot(accuracy_rff, rff_time, label="RFF (Time)", color='tab:cyan', marker='o')
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax1.grid(True, which="both", linestyle="--", linewidth=0.5)

# Ajouter les points pour le modèle de base (accuracy 87%)
ax1.scatter(baseline_accuracy, baseline_time, color='tab:green', marker='o', label="Baseline (Time)")

# Axe droit : Mémoire
ax2 = ax1.twinx()
ax2.set_ylabel("Memory Usage (MiB)", color='tab:red')
ax2.plot(accuracy_nystrom, nystrom_memory, label="Nyström (Memory)", color='tab:red', marker='x')
ax2.plot(accuracy_rff, rff_memory, label="RFF (Memory)", color='tab:orange', marker='x')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Ajouter le point pour la mémoire du modèle de base (accuracy 87%)
ax2.scatter(baseline_accuracy, baseline_memory, color='tab:green', marker='x', label="Baseline (Memory)")

# Ajout de la légende
fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.85), frameon=False)

# Titre et affichage
plt.title("Comparison of Nyström and RFF: Time and Memory vs. Accuracy")
plt.tight_layout()
plt.show()
```
